```python
# Import the datasets and llbraries
from sklearn.datasets.samples_generator import make_blobs

# Useful in creating, training and testing our MACHINE lEARNING Model
from sklearn import svm

# Useful in Visualizing our Data points 
import matplotlib.pyplot as plt

# As the name suggest, Numerical Python helps in carrying out Numerical Operation
import numpy as np

# Generate isotropic Gaussian blobs for clustering

X,Y= make_blobs(n_samples=2000,centers=2,random_state=6)

# Building our MACHINE LEARNING linear Classifier
clf=svm.SVC(kernel='linear',C=1,gamma=1)
clf2=svm.SVC(kernel='poly',C=1,gamma=1)


# Training our model
clf.fit(X,Y)
clf2.fit(X,Y)

# A scaater plot to visualize the data points 
plt.scatter(X[:,0],X[:,1],c=Y,s=12 )

ax=plt.gca()

# Get axes limits
xlim=ax.get_xlim()
ylim=ax.get_ylim()

# Spacing within 30 points
xx=np.linspace(xlim[0],xlim[1],30)
yy=np.linspace(ylim[0],ylim[1],30)

# Create Meshgrid

YY,XX=np.meshgrid(yy,xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T

# Plotting the Hyperplanes
Z = clf.decision_function(xy).reshape(XX.shape)
Z2=clf2.decision_function(xy).reshape(XX.shape)


# Plotting our linear Classifier
ax.contour(XX,YY,Z,levels=[-1,0,1])

# Plotting our Polynomial Classifier
ax.contour(XX,YY,Z2,levels=[-1,0,1])

```
